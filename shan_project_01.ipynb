```python
print("\nNULL VALUES:")
print(df.isnull().sum())
```

```python
print("\nDuplicate Order IDs:", df['order_id'].duplicated().sum())
```

```python
df = df.drop_duplicates(subset='order_id')# just cleaning cloummn based duplicate
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['order_value'] = pd.to_numeric(df['order_value'], errors='coerce')
```

```python
df['restaurant_name'] = df['restaurant_name'].str.strip()
```

```python
df['payment_method'] = df['payment_method'].str.strip()
```

```python
df['discount_used'] = df['discount_used'].astype(str).str.strip()
```

```python
df['discount_used'] = df['discount_used'].str.strip()
```

```python
print("\nOrder Value Stats:")
print(df['order_value'].describe())
```

```python
df = df.reset_index(drop=True)
```

```python
print("\nFinal Shape:", df.shape)
print("\nCleaned Data Sample:")
print(df.head())
```
